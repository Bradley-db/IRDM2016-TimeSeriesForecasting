```python
import pandas as pd
import numpy as np

from datetime import datetime,date
import holidays
```

#  produce all features for purpose of prediction

```python
load_history_long = pd.read_csv('./data/load_history_long.csv')
```

```python
load_history_long.head()
```

```python
load_history_long.shape
```

```python
# take load data out of dataframe
load = load_history_long['load']
del load_history_long['load']
```

```python
# check holiday
us_holidays = holidays.UnitedStates()
```

```python
# create holiday and weekday feature
holiday = []
weekday = []
for i in xrange(load_history_long.shape[0]):
    thisdate = date(load_history_long['year'][i],load_history_long['month'][i], load_history_long['day'][i])
    isholiday = int(thisdate in us_holidays)
    holiday.append(isholiday)
    dayofweek = thisdate.weekday()+1
    weekday.append(dayofweek)
```

```python
# add temperature per hour feature for all stations
```

```python
temp_per_day = pd.read_csv('./data/temp_prediction.csv')
```

```python
temp_per_day.head()
```

```python
temp_per_day.shape
```

```python
# combine all features
```

```python
load_history_long['holiday'] = holiday
load_history_long['weekday'] = weekday
load_history_long['load'] = load
```

```python
load_history_long.head()
load_history_long.shape
```

```python
result = pd.merge(load_history_long, temp_per_day, how='outer', on=['year','month','day','hour'])
```

```python
# put load to the last column
load = result['load']
del result['load']
result['load'] = load
```

```python
result
```

```python
result.to_csv('./data/feature_data.csv', index=None)
```
